```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# NECTA 2024 CSEE Results Index Page
index_url = "https://matokeo.necta.go.tz/results/2024/csee/CSEE2024/CSEE2024.htm"
base_url = "https://matokeo.necta.go.tz/results/2024/csee/CSEE2024/"

# Step 1: Get the page and parse it
res = requests.get(index_url)
soup = BeautifulSoup(res.text, 'html.parser')

# Step 2: Extract all the links to individual school result pages
links = soup.find_all('a')

# Collecting only valid hyperlinks that point to school results (these links end in .htm)
school_links = [base_url + link['href'] for link in links if link.get('href', '').endswith(".htm")]

# Print the first few links to confirm
print(f"Found {len(school_links)} school result links.")
print(school_links[:5])  # Show the first 5 links for checking

# Step 3: Save the school hyperlinks to a CSV file for easy reference
df = pd.DataFrame(school_links, columns=["School_Results_Link"])
df.to_csv("school_links.csv", index=False)

print("\n✅ Hyperlinks saved to school_links.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Load the CSV file containing school result page links
school_links_df = pd.read_csv("school_links.csv")

# Step 2: Initialize a list to store scraped data
school_data = []

# Step 3: Loop through each link to scrape the data from each school result page
for school_link in school_links_df['School_Results_Link']:
    try:
        # Request the school result page
        school_res = requests.get(school_link)
        school_res.raise_for_status()  # Check if the request was successful (status code 200)
        
        school_soup = BeautifulSoup(school_res.text, 'html.parser')
        
        # Attempt to extract relevant information, with better error handling
        school_name = school_soup.find('h1')  # Adjust if 'h1' is not correct
        school_name = school_name.text.strip() if school_name else 'N/A'

        # Example for 'Region' - Adjust according to the actual HTML structure
        region_tag = school_soup.find('td', string='Region')
        region = region_tag.find_next('td').text.strip() if region_tag else 'N/A'

        # Example for 'Total Students' and 'DIVISION ONE' - Adjust according to the actual HTML structure
        total_students_tag = school_soup.find('td', text='Total Students')
        total_students = int(total_students_tag.find_next('td').text.strip()) if total_students_tag else 0
        
        div_one_tag = school_soup.find('td', text='DIVISION ONE')
        div_one_students = int(div_one_tag.find_next('td').text.strip()) if div_one_tag else 0
        
        # Check if the school is private (adjust based on actual page content)
        is_private = "Private" in school_name  # Adjust logic if needed
        
        # Add the data to the list
        school_data.append({
            'School_Name': school_name,
            'Region': region,
            'Total_Students': total_students,
            'DIV_ONE': div_one_students,
            'Is_Private': is_private
        })
        
    except requests.exceptions.RequestException as req_err:
        print(f"Request error for {school_link}: {req_err}")
    except Exception as e:
        print(f"Error scraping {school_link}: {e}")

# Step 4: Convert the scraped data to a DataFrame
df = pd.DataFrame(school_data)

# Step 5: Save the data to a CSV file for further analysis
df.to_csv("school_data.csv", index=False)

# Step 6: Print the first few rows of the data to verify
print(df.head())
```
